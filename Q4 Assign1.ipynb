```python
import pandas as pd

wine_data = pd.read_csv('WineQT.csv')
```

```python
wine_data
```

### convert all the values in the quality attribute to 0 (bad) if the value is less than or equal to 6 and to 1 (good) otherwise

```python
def bin_qual(i):
    if i<=6:
        return 0
    else:
        return 1
    
wine_data['quality'] = wine_data['quality'].apply(bin_qual)
```

```python
wine_data
```

```python
wine_data['quality'].value_counts()
```

```python
wine_data_copy = wine_data.copy()
```

### MIN_MAX Scaling for other attributes leaving 'quality'

### X' = (X - min(x)) / (max(x) - min(x))

```python
for col in wine_data_copy.columns:
    if col != 'quality':
        wine_data_copy[col] = (wine_data_copy[col] - wine_data_copy[col].min()) / (wine_data_copy[col].max() - wine_data_copy[col].min())
```
